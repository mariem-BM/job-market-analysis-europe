```python
import pandas as pd

df = pd.read_csv("clean_jobs_data.csv")
df.head()
```

```python
#create new column = averae of salary_min & salary_max

df["salary_avg"] = (df["salary_min"] + df["salary_max"]) / 2
df[["title", "country", "salary_min", "salary_max", "salary_avg"]].head()
```

```python
df["salary_missing"] = df["salary_avg"].isna() 
# Group by country, count how many True (missing) values per group
df.groupby("country")["salary_missing"].sum()
```

```python
# Loop through each country, print its top 5 highest paid offers
for country in df["country"].unique():
    print(f"\n--- Top 5 highest paid offers in {country} ---")
    top5 = df[df["country"] == country].sort_values("salary_avg", ascending=False).head(5)
    print(top5[["title", "company", "salary_avg"]])
```

```python
skill_columns = [col for col in df.columns if col.startswith("skill_")]

# Calculate skill demand as a percentage of total offers per country
skill_pct = df.groupby("country")[skill_columns].mean() * 100

skill_pct.round(1)
```

```python
# Extract month from the created date column
df["month"] = pd.to_datetime(df["created"]).dt.to_period("M").astype(str)

# Count offers per month per country
df.groupby(["month", "country"])["title"].count().reset_index()
```
